```python
import numpy as np
import pandas as pd
```

# Series 기본

Series는 pandas의 1차원 자료구조이다.

index를 직접 지정하여 원하는 라벨을 붙일 수 있다.

```python
s = pd.Series(
    [12356784, 5437689, 3440451, 2805246],
    index=['Seoul', 'Busan', 'Incheon', 'Daegu']
)
```

Series 자체와, 인덱스에 이름을 붙일 수 있다.

```python
s.name = 'Population'
s.index.name = 'City'
print(s)
```

index는 각 데이터의 라벨 목록이고, values는 실제 값들이다.

```python
print(s.index)
print(s.values)
```

딕셔너리를 시리즈로 만들면 key가 인덱스, value가 값이 된다.

이때 index를 따로 지정하여 원하는 순서대로 만들 수 있다.

```python
data = {
    'Seoul': 12356784,
    'Incheon': 3440451,
    'Busan': 5437689,
    'Daegu': 2805246
}
s = pd.Series(data, index=['Seoul', 'Busan', 'Incheon', 'Daegu'])
s.name = 'Population'
s.index.name = 'City'
print(s)
```

# Series 인덱싱

iloc은 위치 기반 인덱싱이고, loc은 라벨 기반 인덱싱이다.

```python
print(s.iloc[1])
print(s.loc['Busan'])
```

여러 개를 고를 때는 리스트를 사용한다.

```python
print(s.iloc[[0, 3, 1]])
print(s.loc[['Seoul', 'Daegu', 'Busan']])
```

슬라이싱도 가능하다. iloc은 끝 번호를 포함하지 않고, loc은 끝 라벨도 포함하는 차이가 있다.

```python
print(s.iloc[1:3])
print(s.loc['Busan':'Daegu'])
```

조건식을 사용해서 원하는 값만 고를 수 있다.

```python
print(s[(300e4 < s) & (s < 500e4)])
```

# Series 연산

numpy처럼 벡터화 연산이 가능하다.

```python
print(s / 1_000_000)
```

시리즈끼리 연산하면 같은 인덱스끼리 계산된다.

한쪽에만 있는 인덱스는 계산할 수 없어 NaN이 된다.

```python
s1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
s2 = pd.Series([25, 35, 45], index=['b', 'c', 'd'])
s2 - s1
```

notnull() 함수로 NaN이 값만 볼 수 있다.

```python
s3 = s2 - s1
s3[s3.notnull()]
```

Series.values끼리 연산하면 같은 순번에 해당하는 데이터끼리 처리된다.

```python
s2.values - s1.values
```
