```python
import numpy as np
import pandas as pd
```

# DataFrame 기본

DataFrame은 표 형태의 2차원 자료구조이다.

index, columns를 직접 지정하여 원하는 라벨을 붙일 수 있다.

딕셔너리를 사용하면 key를 컬럼 라벨로 사용한다.

```python
data = {
    '2022': [12350000, 5437000, 3440200, 2805000],
    '2023': [12352000, 5437200, 3440300, 2805050],
    '2024': [12353000, 5437400, 3440400, 2805100],
    '2025': [12356784, 5437689, 3440451, 2805246],
    '지역': ['수도권', '경상권', '수도권', '경상권'],
    '2015~2019 증가율': [0.0283, 0.0163, 0.0982, 0.0141]
}
df = pd.DataFrame(data,
                  index=['서울', '부산', '인천', '대구'],
                  columns=['지역', '2022', '2023', '2024', '2025', '2015~2019 증가율'])
print(df)
```

인덱스 라벨과 컬럼 라벨에 이름을 붙일 수 있다.

```python
df.index.name = '도시'
df.columns.name = '특성'
print(df)
```

index는 행 라벨, columns는 열 라벨, values는 실제 값들이다.

```python
print(df.index)
print(df.columns)
print(df.values)
```

없는 컬럼에 대입하면, 새 컬럼으로 추가한다.

```python
df['2022~2025 증가율'] = ((df['2025'] - df['2022']) / df['2022'] * 100).round(2)
print(df)
```

del 키워드로 컬럼을 삭제할 수 있다.

```python
del df['2022~2025 증가율']
print(df)
```

인덱스나 컬럼 라벨을 지정하지 않으면 0, 1, 2, ... 같은 정수 인덱스를 자동으로 붙인다.

```python
pd.DataFrame(np.arange(12).reshape(3, 4))
```

# DataFrame 인덱싱

컬럼 라벨을 사용해 컬럼을 인덱싱할 수 있다.

한 컬럼만 고르면 결과는 Series이다.

여러 컬럼을 고르면 결과는 DataFrame이다. 

```python
df['지역']
```

```python
df[['2022', '2023']]
```

loc을 사용해 인덱스 라벨 기반으로 인덱싱 할 수 있다.

```python
df.loc['서울']
```

```python
df.loc[['부산', '대구']]
```

```python
df.loc['부산':'대구']
```

```python
df.loc[df['2023'] > 5_000_000, ['지역', '2025']]
```

iloc을 사용해 정수 인덱스 기반으로 인덱싱할 수 있다.

```python
df.iloc[-1]
```

```python
df.iloc[[2, 0]]
```

```python
df.iloc[1:3]
```

인덱스에 대해 슬라이싱을 사용할 수 있다. 정수 인덱스와 달리 인덱스 라벨을 사용하면 끝 라벨도 포함한다.

```python
print(df[1:3])
print(df['부산':'대구'])
```

컬럼 라벨과 인덱스 라벨을 모두 사용해 원소 하나만 고를 수 있다.

```python
df['2025']['서울']
```

컬럼 라벨을 지정하지 않은 경우에만 정수 인덱싱을 사용할 수 있다.

```python
df = pd.DataFrame(np.arange(12).reshape(3, 4))
df[2]
```

# DataFrame 인덱스 조작

```python
df = pd.DataFrame(
        np.vstack([list("ABCDE"), np.round(np.random.rand(3, 5), 2)]).T,
        columns=["c1", "c2", "c3", "c4"],
    )
df
```

set_index()는 기존 컬럼 하나를 인덱스 라벨로 바꾼다.

```python
df2 = df.set_index('c1')
df2
```

reset_index()는 인덱스 라벨을 없애고 컬럼으로 추가한다.

```python
df2.reset_index()
```

drop을 True로 하면 컬럼으로 추가하지 않고 버린다.

```python
df2.reset_index(drop=True)
```

Multi index는 인덱스나 컬럼에 계층 구조를 만든다.

```python
df = pd.DataFrame(
    np.round(np.random.randn(6, 4), 2),
    columns=[["A", "A", "B", "B"], ["C1", "C2", "C3", "C4"]],
    index=[["M", "M", "M", "F", "F", "F"], [f"id_{i+1}" for i in range(3)] * 2]
)
df.columns.names = ["Cidx1", "Cidx2"]
df.index.names = ["Ridx1", "Ridx2"]
print(df)
```

stack()은 컬럼을 인덱스로 쌓는다.

```python
df.stack('Cidx1')
```

unstack()은 인덱스를 컬럼으로 쌓는다.

```python
df.unstack('Ridx2')
```

# DataFrame 연산

컬럼 라벨로 인덱싱 하여 컬럼끼리 연산할 수 있다.

없는 컬럼 라벨에 대입하면, 새로운 컬럼을 추가한다.

```python
data = {
    'math': [90, 80, 70],
    'english': [85, 95, 75],
    'science': [90, 85, 80]
}
df = pd.DataFrame(data)

df['total'] = df['math'] + df['english'] + df['science']
```

# DataFrame 집계

NaN은 결측값(값이 없음)을 의미한다.

count()로 NaN을 제외한 개수를 셀 수 있다.

```python
s = pd.Series(range(10), dtype=float)
s[3] = np.nan
s.count()
```

```python
df = pd.DataFrame(np.random.randint(5, size=(4, 4)), dtype=float)
df.iloc[2, 3] = np.nan
df.count()
```

sort_values()는 값을 기준으로 정렬한다.

ascending이 True면 오름차순 정렬한다.

by는 지정한 컬럼을 정렬의 기준으로 삼는다.

```python
s.sort_values(ascending=False)
```

```python
df.sort_values(by=2)
```

value_counts()는 각 값이 몇 번 나타났는지 센다.

```python
s = pd.Series(np.random.randint(6, size=100))
s.value_counts()
```

sort_index()는 인덱스를 기준으로 정렬한다.

```python
s.value_counts().sort_index()
```

sum()은 axis=0이면 열 방향, axis=1이면 행 방향으로 합을 계산한다.

```python
df = pd.DataFrame(np.random.randint(10, size=(4, 8)))
df
```

```python
df.sum()
```

```python
df.sum(axis=1)
```

apply()를 사용하면 행 또는 열 단위로 사용자 정의 함수를 적용할 수 있다.

axis=0이면 기본인 열에, axis=1이면 행에 적용한다.

```python
df = pd.DataFrame({
    'A': [1, 3, 1, 3, 4],
    'B': [2, 3, 1, 2, 3],
    'C': [1, 3, 2, 4, 4],
})
df
```

```python
df.apply(lambda x: x.max() - x.min())
```

```python
df.apply(lambda x: x.max() - x.min(), axis=1)
```

```python
freq_table = df.apply(lambda x: x.value_counts(), axis=0)
freq_table
```

fillna()는 NaN을 원하는 값으로 채운다.

```python
freq_table.fillna(0.0).astype(int)
```

cut()은 숫자 데이터를 구간으로 나누어 범주형 데이터로 바꾼다.

```python
ages = [0, 2, 10, 21, 23, 37, 31, 61, 20, 41, 80, 101]
bins = [1, 20, 30, 50, 70, 100]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
result = pd.cut(ages, bins=bins, labels=labels)
print(result)
print("categories:", result.categories)
print("codes:", result.codes)
```

# DataFrame 파일 입출력

데이터프레임을 파일로 저장하고, 불러오는 기능을 제공한다.

불러오면서 Unnamed: 0이 생기는 문제를 해결하기 위해 read_csv 함수에서 index_col을 0으로 지정했다.

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [20, 25, 30],
    'city': ['Seoul', 'Incheon', 'Busan']
}

df = pd.DataFrame(data)
df.to_csv('test.csv', mode='w', sep=',')

pd.read_csv('test.csv', index_col=0)
```
